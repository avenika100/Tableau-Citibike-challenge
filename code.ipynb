```python
import pandas as pd
import datetime
```

```python
# Read data
citi_bike_jan = pd.read_csv('JC-202501-citibike-tripdata.csv')
citi_bike_feb = pd.read_csv('JC-202502-citibike-tripdata.csv')
citi_bike_march = pd.read_csv('JC-202503-citibike-tripdata.csv')
```

```python
#create data frame for citi bike data January 
citibike_jan_df = pd.DataFrame(citi_bike_jan)
citibike_jan_df.tail(5)
```

```python
#create data frame for citi bike data February  
citibike_feb_df = pd.DataFrame(citi_bike_feb)
citibike_feb_df.tail(5)
```

```python
#create data frame for citi bike data March
citibike_march_df = pd.DataFrame(citi_bike_march)
citibike_march_df.head(5)
```

```python
#Merging Jan and Feb data
first_merge = pd.concat([citibike_jan_df, citibike_feb_df], ignore_index=True)
first_merge
```

```python
#final merge
final_merge = pd.concat([first_merge, citibike_march_df], ignore_index=True)
final_merge
```

```python
#drop duplicates 
final_merge = final_merge.drop_duplicates()
```

```python
final_merge.columns
```

```python
#checking for missing values
final_merge.isnull
```

```python
#change to date time objects
final_merge['started_at'] = pd.to_datetime(final_merge['started_at'])
final_merge['ended_at'] = pd.to_datetime(final_merge['ended_at'])
```

```python
#remove invalid or impossible data
final_merge = final_merge[final_merge['ended_at'] > final_merge['started_at']]
final_merge['trip_duration_min'] = (final_merge['ended_at'] - final_merge['started_at']).dt.total_seconds() / 60
final_merge = final_merge[(final_merge['trip_duration_min'] >= 1) & (final_merge['trip_duration_min'] <= 180)]

final_merge
```

```python
#create hour of day field and day of week field
final_merge['day_of_week'] = final_merge['started_at'].dt.day_name()
final_merge['hour_of_day'] = final_merge['started_at'].dt.hour

final_merge
```

```python
#create month of year field 
final_merge['month'] = final_merge['started_at'].dt.strftime('%B')
final_merge
```

```python
#create new data frame for the cleaned data
cleaned_winter_citibike_data = pd.DataFrame(final_merge)
cleaned_winter_citibike_data
```

```python
#save the  dataframe to csv
cleaned_winter_citibike_data.to_csv('cleaned_winter_citibike_data.csv', index=False)
```
